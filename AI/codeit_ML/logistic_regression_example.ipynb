```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

import pandas as pd
```

## 데이터 준비

```python
iris_data = load_iris()
```

```python
print(iris_data.DESCR)
```

```python
X = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)

X
```

```python
y = pd.DataFrame(iris_data.target, columns=['class'])

y
```

## 로지스틱 회귀

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
y_train = y_train.values.ravel()  # logistic_regression을 진행할 때 추가할 것을 추천
```

```python
model = LogisticRegression(solver='saga', max_iter=2000) # solver: 모델 최적화시 어떤 알고리즘을 사용할 것인가 # max_iter: 최적화를 몇 번 진행할 것인가(최대)
```

```python
model.fit(x_train, y_train)
```

```python
y_prediction = model.predict(x_test)
```

```python
y_prediction
```

```python
model.score(x_test, y_test)  # 예측한 것중에 몇 %가 올바르게 분리됐는지를 계산해주는 함수
```
