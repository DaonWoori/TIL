## 데이터 불러오기

```python
data_path = "/content/drive/MyDrive/colab/codeit/data/"
```

```python
titanic_df = pd.read_csv(data_path + 'titanic.csv')
```

```python
titanic_df.head()
```

## one-hot encoding

```python
titanic_sex_embarked = titanic_df[['Sex', 'Embarked']]
```

```python
titanic_sex_embarked.head()
```

```python
# 가장 쉬운 방법
# 변환하고자하는 컬럼만 넘겨주는 방법
one_hot_encoded_df = pd.get_dummies(titanic_sex_embarked)
```

```python
one_hot_encoded_df.head()
```

```python
# optional parameter를 활용한 방법
# 전체 데이터를 넘겨주고 변환하길 원하는 컬럼만 변환하는 방법

one_hot_encoded_df = pd.get_dummies(data=titanic_df, columns=['Sex', 'Embarked'])
one_hot_encoded_df.head()
```
